**1. 책을 구조체로 표현하자. 책에 대한 정보는 제목(문자열, 최대 100문자), 저자(문자열, 최대 100문자), 장르(열거형) 등으로 구성된다. 책의 장르는 COMIC, SF, DOCU, DRAMA 중의 하나로 분류하고 열거형으로 구현한다. {"바람과 함께 사라지다", "마가렛 미첼", DRAMA }의 값을 가지는 구조체 변수를 생성했다가 다시 화면에 출력해보자.**

```c
#include <stdio.h>
typedef enum {
    COMIC,
    SF,
    DOCU,
    DRAMA,
} GENRE;
 
//enum to string
const char* genreToString(GENRE genre){
    const char* table[] = {
        "COMIC",
        "SF",
        "DOCU",
        "DRAMA"
    };
    return table[genre];
}
struct book{
    char title[100];
    char author[100];
    GENRE type;
};

void printBook(const struct book* book){
    printf("{ %s, %s, %s }", book->title, book->author, genreToString(book->type));
}

//book 구조체를 생성하여 출력하는 프로그램.
int main(void){
    struct book bookObject = {
        .title = "바람과 함께 사라지다",
        .author = "마가렛 미첼",
        .type = DRAMA
    };
    printBook(&bookObject);
    return 0;
}
```

**2. 1번 문제에서 정의한 2개의 구조체를 받아서 책의 저자가 일치하는지를 검사하는 함수 equal_author()을 작성해보자. equal_author()은 다음과 같은 원형을 가진다. b1과 b2의 저자가 같으면 1을 반환하고 그렇지 않으면 0을 반환한다. 함수를 작성하여 테스트한다.**

```c
#include <stdio.h>
#include <string.h>
typedef enum {
    COMIC,
    SF,
    DOCU,
    DRAMA,
} GENRE;

//enum to string
const char* genreToString(GENRE genre){
    const char* table[] = {
        "COMIC",
        "SF",
        "DOCU",
        "DRAMA"
    };
    return table[genre];
}
struct book{
    char title[100];
    char author[100];
    GENRE type;
};

//strcmp function은 두 string이 같을때만 0을 반환한다.
int equal_author(struct book b1, struct book b2){
    return(strcmp(b1.author, b2.author) == 0);
}

void printBook(const char* name, const struct book* book){
    printf("%s = { %s, %s, %s }\n",name, book->title, book->author, genreToString(book->type));
}

//두 책의 저자가 같은지 비교하여 출력하는 프로그램.
int main(){
    struct book b1 = {
        .title = "노인과 바다",
        .author = "헤밍웨이",
        .type = DRAMA
    };
    struct book b2 = {
        .title = "누구를 위하여 종을 울리나",
        .author = "헤밍웨이",
        .type = DRAMA
    };

    printBook("b1", &b1);
    printBook("b2", &b2);
    printf("equal_author()의 반환값: %d", equal_author(b1, b2));
    return 0;
}
```

**4. 구조체를 이용하여 복소수를 정의하고 복소수의 덧셈을 수행하는 함수를 작성하고 테스트하시오.**

```c
#include <stdio.h>
struct complex{
    double real;
    double imag;
};

//복소수를 더하는 함수
struct complex complex_add(struct complex c1, struct complex c2){
    c1.real += c2.real;
    c1.imag += c2.imag;
    return c1;
}

//복소수 개체를 출력하는 함수
void printComplex(struct complex c){
    printf("%.2lf+%.2lfi\n", c.real, c.imag);
}

int main(){
    struct complex c1 = {
        .real = 1.00,
        .imag = 2.00
    };
    struct complex c2 = {
        .real = 2.00,
        .imag = 3.00
    };

    printComplex(c1);
    printComplex(c2);
    printComplex(complex_add(c1, c2));

    return 0;
}
```

**5. 2차원 평면에서 점은 (x, y) 좌표로 나타낼 수 있다. point 구조체를 받아서 두 점의 좌표가 일치하면 1을 반환하고 그렇지 않으면 0을 반환하는 함수를 작성하고 테스트하라.**

```c
#include <stdio.h>
typedef struct{
    int x, y;
}point;

//두 point 구조체가 같은지 다른지 비교하는 함수
int equal(point a, point b){
    return a.x == b.x && a.y == b.y;
}


//point 구조체를 출력하는 함수
void print_point(const point *p){
    printf("(%d, %d)", p->x, p->y);
}

int main(void){
    point a = {
        .x = 1,
        .y = 2
    };
    point b = {
        .x = 3,
        .y = 5
    };

    print_point(&a);
    printf(" %s ", equal(a, b) ? "==" : "!=");
    print_point(&b);
    
    return 0;
}
```

**7. 2차원 공간에 있는 점의 좌표를 받아서 이 점이 속하는 사분면의 번호를 반환하는 함수를 작성하고 테스트하라.**

```c
#include <stdio.h>
typedef struct{
    int x, y;
}point;

//좌표가 속한 사분면을 반환하는 함수
int quadrant(point p){
    if(p.x > 0 && p.y > 0) return 1;
    else if(p.x > 0 && p.y < 0) return 4;
    else if(p.x < 0 && p.y < 0) return 3;
    else return 2;
}

//point 구조체를 출력하는 함수
void print_point(const point *p){
    printf("(%d, %d)", p->x, p->y);
}

int main(void){
    point a = {
        .x = -1,
        .y = 2
    };
    print_point(&a);
    printf("의 사분면 = %d", quadrant(a));
    return 0;
}
```

**8. 원을 나타내는 circle 구조체를 정의하고, 원의 면적을 계산하는 함수와 원의 둘레를 계산하는 함수를 작성하라.**

```c
#include <stdio.h>
typedef struct{
    int x,y;
}point;

typedef struct{
    point center;
    double radius;
}CIRCLE;

//원의 면적과 둘레를 구하는 함수
const double pi = 3.14;
double area(CIRCLE c){
    return c.radius * c.radius * pi;
}

double perimeter(CIRCLE c){
    return c.radius * 2 * pi;
}

int main(void){
    CIRCLE a = {
        .center = {
            .x = 0,
            .y = 0
        },
        .radius = 10
    };

    printf("원의 중심점: (%d, %d)\n", a.center.x, a.center.y);
    printf("원의 반지름: %.0lf\n", a.radius);
    printf("원의 면적: %.2lf\n",area(a));
    printf("원의 둘레: %.2lf\n",perimeter(a));

    return 0;
}
```

**10. 직원을 나타내는 구조체 employee가 사번, 이름, 전화번호, 나이 등으로 구성되어 있다. 구조체의 배열을 선언하고 10명의 데이터로 초기화하라. 이 중에서 나이가 20 이상 30 이하인 직원을 찾아서 출력하도록 하라. typedef을 사용하여서 구조체 타입을 정의해서 사용하라.**

```c
#include <stdio.h>
#include <string.h>
#include <stdlib.h>
#include <time.h>
const int length = 10;

typedef struct{
    int id;
    char name[30];
    char tel[30];
    int age;
}employee;

//랜덤한 데이터로 배열을 채우는 함수
void fill_array(employee* lst, int len){
    srand(time(NULL));
    for(int i = 0; i < len; i++){
        lst[i].id = i+1;
        sprintf(lst[i].name, "홍길동%d", i+1);
        strcpy(lst[i].tel, "010-4816-0523");
        lst[i].age = (rand()%35) + 1;
    }
}

int main(void){
    employee employeeList[length];
    fill_array(employeeList, length);

    for(int i = 0; i < length; i++){
        int age = employeeList[i].age;
        if(age >= 20 && age <= 30) 
            printf("이름 : %s 나이=%d\n", employeeList[i].name, employeeList[i].age);
    }

    return 0;
}
```

**13. 삼각형, 사각형, 원을 동시에 표현 할 수 있는 공용체를 설계하라. 사용자로부터 도형의 종류와 도형의 데이터를 받아서 저장해보자.**

```c
#include <stdio.h>
struct shape {
    int type;
    union{
        struct {int base, height; } tri;
        struct {int width, height; } rect;
        struct {int radius; } circ;
    } data;
};
typedef struct shape shape;

int main(void){
    shape shapeObject;

    printf("도형의 타입을 입력하시오(0, 1, 2): ");
    scanf("%d", (&shapeObject.type));

    //데이터 입력
    switch(shapeObject.type){
        case 0:
            printf("밑변과 높이를 입력하시오: ");
            scanf("%d %d", &(shapeObject.data.tri.base), &(shapeObject.data.tri.height));
            break;
        case 1:
            printf("가로와 세로를 입력하시오: ");
            scanf("%d %d", &(shapeObject.data.rect.width), &(shapeObject.data.rect.height));
            break;
        case 2:
            printf("반지름을 입력하시오");
            scanf("%d", &(shapeObject.data.circ.radius));
            break;
        default:
            printf("올바르지 않은 타입");
            return 0;
    }

    //면적 계산과 출력
    int area;
    switch(shapeObject.type){
        case 0:
            area = shapeObject.data.tri.base * shapeObject.data.tri.height / 2;
            break;
        case 1:
            area = shapeObject.data.rect.width * shapeObject.data.rect.height;
            break;
        case 2:
            area = shapeObject.data.circ.radius * shapeObject.data.circ.radius * 3.14;
            break;
    }

    printf("면적은 %d", area);
    return 0;
}
```

**14. 데이터베이스의 기능을 하는 간단한 프로그램을 작성하여 보자. 이 프로그램은 mp3와 같은 음악 파일을 관리한다. 사용자는 음악 파일을 추가, 삭제, 출력 할 수 있으며 제목을 가지고 특정 곡을 탐색 할 수 있다.**

```c
#include <stdio.h>
#include <string.h>
#define MAX_STRING_SIZE 100

//char[] type-alias.
typedef char string[MAX_STRING_SIZE];

typedef struct{
    string title;
    string artist;
    string directory;
    short genre;
}music;

const char* genreList[] = {
    "가요",
    "팝",
    "클래식",
    "영화음악"
};

int idx = 0;
music musicList[100];

void flush_stdin(){
    /*
    scanf()를 사용하여 stdin에서 입력을 받은 후, 버퍼를 비우지 않으면
    개행 문자가 버퍼에 남아있어 fgets()가 제대로 동작하지 않을 수 있습니다.
    이러한 문제를 막기 위하여 fgets()를 사용하기 전에 반드시 stdin 버퍼를 비우십시오.

    NOTE: fflush(stdin)을 사용하지 마십시오.
    fflush()는 출력 버퍼를 비우는 함수입니다. fflush()에 stdin을 넣는 것은,
    함수 설계 의도에 위반됩니다.
    
    fflush()가 입력 버퍼에서도 동작 하는 것은 POSIX C 확장이며, 
    이는 C의 표준이 아니므로 모든 환경에서 적절하게 동작하지 않을 수 있습니다.
    */
    while ( getchar() != '\n' );
}

void get_string(char* dest){
    string input;
    int len;
    fgets(input, MAX_STRING_SIZE, stdin);
    len = strlen(input);

    /*
    NOTE: fgets()로 받은 문자열의 마지막 문자는 개행 문자일 수도 있습니다.
    */
    if(input[len-1] == '\n') input[len-1] = 0;
    strcpy(dest, input);
}

void add_music(){
    //issue: fgets() is not supported in jupyter-c-kernel.
    //fgets()를 사용하기 전에 stdin 버퍼를 비우십시오.
    flush_stdin();

    printf("제목: ");
    get_string(musicList[idx].title);
    printf("가수: ");
    get_string(musicList[idx].artist);
    printf("위치: ");
    get_string(musicList[idx].directory);

    int genre;
    printf("장르(0: 가요, 1:팝, 2:클래식, 3:영화음악): ");
    scanf("%d", &genre);

    //입력 값 검증
    if(genre < 0 || genre > 3){
        printf("올바르지 않은 장르.");
        return;
    }
    musicList[idx].genre = genre;
    idx++;
}

void print_music(int index){
    if(index > idx) return;
    printf("%s - %s\n", musicList[index].artist, musicList[index].title);
    printf("%s\n", genreList[musicList[index].genre]);
    printf("%s\n\n", musicList[index].directory);
}

void show_music(){
    if(idx == 0){
        printf("저장 된 음악이 없습니다.");
        return;
    }

    for(int i = 0; i < idx; i++){
        printf("(%d / %d)\n", i+1, idx);
        print_music(i);
    }
}

void search_music(){
    string query;
    printf("검색 할 음악의 제목을 입력하십시오: ");

    flush_stdin();
    get_string(query);

    for(int i = 0; i < idx; i++){
        if(strcmp(musicList[i].title, query)  == 0){
            print_music(i);
            return;
        }
    }

    printf("해당 음악을 찾을 수 없습니다.");
}

int main(void){
    const char* separator = "=========================\n";

    //함수 포인터 배열
    void(*commands[])() = {
        &add_music,
        &show_music,
        &search_music
    };

    int menu;
    while(1){
        //issue: puts() via stdio-wrap in jupyter-c-kernel may not work properly.
        printf("%s", separator);
        printf("1. 추가\n2. 출력\n3. 검색\n4. 종료\n");
        printf("%s", separator);
        printf("매뉴를 선택하십시오: ");
        scanf("%d", &menu);

        if(menu > 0 && menu < 4){
            commands[menu-1]();
            printf("\n");
        }
        else if(menu == 4){
            printf("종료합니다.");
            return 0;
        }
        else{
            printf("올바르지 않은 입력.\n\n");
            continue;
        }
    }
}
```

**15. 도서관에서 소장 도서들을 관리할 수 있는 프로그램을 작성하여보자. 책 정보를 저장할 수 있는 구조체를 정의한다. 다음과 같은 메뉴를 생성한 후 각 메뉴 항목을 구현한다.**

```c
#include <stdio.h>
#include <string.h>
#include <stdbool.h>
#include <stdlib.h>
#include <time.h>
#define MAX_STRING_SIZE 100

typedef char string[MAX_STRING_SIZE];

typedef struct{
    int id;
    string title;
    string author;
    bool isCheckouted;
}book;

book bookList[100];
int idx = 0;


void flush_stdin(){
    while ( getchar() != '\n' );
}

void get_string(string dest){
    string input;
    int len;
    fgets(input, MAX_STRING_SIZE, stdin);
    len = strlen(input);

    //개행문자 제거
    if(input[len-1] == '\n') input[len-1] = 0;
    strcpy(dest, input);
}

//리스트에 요소가 있는지 검증
bool check_lst(){
    if(idx == 0){
        puts("데이터베이스에 아무런 책이 없습니다.\n");
        return false;
    }
    else return true;
}

void print_book(int idx){
    book object = bookList[idx];
    printf("%s - %s\n", object.title, object.author);
    printf("ID : %d\n", object.id);
    printf("대출 여부: %s\n\n", object.isCheckouted ? "이미 대출됨." : "대출 가능");
}

void checkout_book(int idx){
    book* object = &bookList[idx];
    if(object->isCheckouted)
        printf("이 도서는 현재 대출중입니다. 반납 처리하시겠습니까? (Y/N): ");
    else
        printf("이 도서는 현재 대출 가능합니다. 대출 처리하시겠습니까? (Y/N): ");
    
    char input;
    flush_stdin();
    scanf("%c", &input);

    if(input == 'Y' || input == 'y')
    {
        object->isCheckouted = !(object->isCheckouted);
        printf("%s 처리 되었습니다.\n\n", object->isCheckouted ? "대출" : "반납");
    }
    else printf("\n");
}

void query_by_id(){
    if(!check_lst()) return;

    int id;
    printf("도서 ID를 입력하십시오: ");
    scanf("%d", &id);

    if(id == 0){
        puts("잘못 된 도서 ID.\n");
        return;
    }
    for(int i = 0; i < idx; i++){
        if(bookList[i].id == id){
            print_book(i);
            checkout_book(i);
            return;
        }
    }

    puts("해당 도서를 찾을 수 없습니다.\n");
}

void query_by_author(){
    if(!check_lst()) return;

    string author;
    bool isExist = false;
    printf("찾을 책의 저자을 입력하십시오: ");
    flush_stdin();
    get_string(author);

    for(int i = 0; i < idx; i++){
        if(strcmp(bookList[i].author, author) == 0){
            isExist = true;
            print_book(i);
        }
    }

    if(!isExist)puts("해당 도서를 찾을 수 없습니다.\n");
}

void query_by_title(){
    if(!check_lst()) return;

    string title;
    printf("찾을 책의 제목을 입력하십시오: ");
    flush_stdin();
    get_string(title);

    for(int i = 0; i < idx; i++){
        if(strcmp(bookList[i].title, title) == 0){
            print_book(i);
            return;
        }
    }

    puts("해당 도서를 찾을 수 없습니다.\n");
}

bool vaildation_data(book target){
    if(idx == 0) return true;

    for(int i = 0; i < idx; i++){
        if(strcmp(target.title, bookList[i].title) == 0) return false;
        if(target.id == bookList[i].id) return false;
    }

    return true;
}

int create_id(){
    while(true){
        srand(time(NULL));
        int id = rand() % 1000 + 1;

        if(idx == 0) return id;
        for(int i = 0; i < idx; i++){
            if(bookList[idx].id == id) continue;
        }
        return id;
    }
}

void add_book(){
    int length;
    book object;
    printf("몆 권의 책을 추가하시겠습니까: ");
    scanf("%d", &length);

    if(length < 0){
        puts("잘못 된 입력.");
        return;
    }
    
    printf("\n");
    flush_stdin();
    for(int i = 0; i < length; i++){
        printf("(%d/%d)\n", i+1, length);
        printf("제목: ");
        get_string(object.title);
        printf("저자: ");
        get_string(object.author);
        object.isCheckouted = false;
        object.id = create_id();

        if(!vaildation_data(object)){
            puts("데이터 검증 오류.\n");
            continue;
        }
        else{
            bookList[idx] = object;
            idx++;
            printf("추가되었습니다. (ID: %d)\n\n", object.id);
        }
    }
}

void show_books(){
    if(!check_lst()) return;
    printf("현재 총 %d 권의 책이 있습니다.\n\n", idx);
}

//list of commandHandlers.
void(*commands[])() = {
    &query_by_id,
    &query_by_author,
    &query_by_title,
    &add_book,
    &show_books
};

int main(void){
    const char* separator = "====================================";
    int menu;
    puts("도서 정보 시스템에 오신 것을 환영합니다.\n");
    while(true){
        puts(separator);
        puts("1. 도서 번호로 책 찾기 (대출 / 반납)");
        puts("2. 저자 이름으로 책 찾기");
        puts("3. 제목으로 책 찾기");
        puts("4. 새로운 책 추가");
        puts("5. 도서관이 소장한 도서의 수 표시");
        puts("6. 종료");
        puts(separator);
        printf("메뉴를 선택하십시오: ");
        scanf("%d", &menu);

        if(menu < 0 || menu > 6){
            puts("올바르지 않은 메뉴.\n");
            continue;
        }
        else if(menu == 6){
            puts("종료합니다.");
            return 0;
        }

        commands[menu-1]();
    }
}
```

**16. 가장 평점이 높은 학생을 찾아서 학생의 이름과 학번, 평점을 화면에 출력하는 프로그램을 작성하시오.**

```c
#include <stdio.h>
#define MAX_STRING_LENGTH 100

typedef char string[MAX_STRING_LENGTH];

typedef struct{
    string id;
    string name;
    double gpa;
}student;

//static binding
student studentList[] = {
    {
        .id = "20233063",
        .name = "손유찬",
        .gpa = 4.13
    },
    {
        .id = "20233066",
        .name = "임해주",
        .gpa = 3.71
    },
    {
        .id = "20210523",
        .name = "노하민",
        .gpa = 1.16
    }

};

//가장 높은 평점의 학생 찾기
int find_max_idx(const student* lst, int size){
    int idx = 0;
    for(int i = 0; i < size; i++){
        if(lst[idx].gpa < lst[i].gpa){
            idx = i;
        }
    }
    return idx;
}

int main(void){
    int max = find_max_idx(studentList,3);
    printf("평점이 가장 높은 학생은 (이름: %s, 학번: %s, 평점: %.2lf)입니다", 
    studentList[max].name, studentList[max].id, studentList[max].gpa );

    return 0;
}
```
